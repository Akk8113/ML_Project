```python
# Required libraries for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

# Settings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

## Load Data Set

```python
# Load the CSV file
df = pd.read_csv("wafer_data.csv")
```

## Basic Info & Structure

```python
# Basic overview
print("Shape of dataset:", df.shape)
print("\nInfo:")
df.info()

# Preview the first few rows
df.head()
```

## Missing Values Analysis


```python
# Calculate missing values
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

# Create summary DataFrame
missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Missing Percentage (%)': missing_percent
})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values(by='Missing Percentage (%)', ascending=False)

# Show top 10 columns with missing data
missing_df.head(10)
```

##  Target / Class Column Analysis

```python
# Target value counts
df['Good/Bad'].value_counts()

# Class imbalance plot
sns.countplot(x='Good/Bad', data=df)
plt.title("Class Distribution")
plt.show()
```

## Statistical Summary

```python
# Statistical summary of numerical features
df.describe().T
```

##  Duplicate Rows Detection

```python
# Check number of duplicate rows
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

##  Correlation Matrix & Heatmap

```python
# Compute correlation with the target
numerics = df.select_dtypes(include=[np.number])
top_corr = numerics.corr().abs().nlargest(20, 'Good/Bad')['Good/Bad'].index

# Plot heatmap of top correlated features
plt.figure(figsize=(12, 10))
sns.heatmap(df[top_corr].corr(), annot=True, cmap='coolwarm')
plt.title("Top 20 Correlated Features with Target")
plt.show()
```

## Outlier Detection (Boxplots)

```python
# Boxplots for outlier detection in selected sensors
selected_sensors = ['Sensor-1', 'Sensor-2', 'Sensor-3']

plt.figure(figsize=(15, 5))
for i, sensor in enumerate(selected_sensors):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x=df[sensor])
    plt.title(f"Outliers in {sensor}")
plt.tight_layout()
plt.show()
```

## Distribution Plots (Histogram/KDE)

```python
# Distribution of selected features
df[selected_sensors].hist(bins=20, figsize=(12, 6))
plt.suptitle("Distribution of Selected Sensor Features")
plt.show()
```

## Boxplots Across Class Labels

```python
# Boxplots of features across Good/Bad classes
plt.figure(figsize=(15, 5))
for i, sensor in enumerate(selected_sensors):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='Good/Bad', y=sensor, data=df)
    plt.title(f"{sensor} vs Good/Bad")
plt.tight_layout()
plt.show()
```

## Class Imbalance Visualization


```python
# Bar plot for class distribution
sns.countplot(x='Good/Bad', data=df)
plt.title("Class Distribution (Good vs Bad Wafers)")
plt.show()
```

## Pairplot for Multivariate Relationships

```python
# Multivariate analysis using pairplot
subset = df[['Sensor-1', 'Sensor-2', 'Sensor-3', 'Good/Bad']]
sns.pairplot(subset, hue='Good/Bad')
plt.show()
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.0)
selector.fit(df.select_dtypes(include=[np.number]))

low_variance_cols = df.select_dtypes(include=[np.number]).columns[~selector.get_support()]
print("Low variance columns:", low_variance_cols.tolist())
```

```python
from sklearn.ensemble import RandomForestClassifier

# Prepare data
X = df.drop(columns=['Unnamed: 0', 'Good/Bad'])
y = df['Good/Bad']

# Handle missing values
X = X.fillna(X.mean())

# Train a random forest
model = RandomForestClassifier()
model.fit(X, y)

# Get importance
importances = pd.Series(model.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(20)

# Plot
top_features.plot(kind='barh', figsize=(10, 8))
plt.title("Top 20 Feature Importances")
plt.gca().invert_yaxis()
plt.show()
```
